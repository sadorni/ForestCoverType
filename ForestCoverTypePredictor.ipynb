The aim of the assignment is to correctly identify the type of forest cover using cartographic informations only. 
The problem is defined by 54 variables both numerical and categorical (represented by binary values) and the dataset comprises of 581012 samples. There are 6 forest cover type. 
We therefore have a large dataset containing a lot of examples on which we want to perform a multi-class classification task. 

We will first **explore** the dataset to see: 
* how the features look for the different forset cover types
* if the dataset is balanced for all the classes
* if there is a need to clean the data, i.e remove features or instances, remove outliers, etc..

We will then **preprocess** the available data and finally we will test different ML algortihms to try and **predict** the cover type: 

* Random Forest Tree 
* Boosted Decision Tree
* KNN
* Support Vector Classifier
* Logistic Regression
* Deep Neural Network
* Naive Bayes Classifier
* Ensemble method using voting 
* Ensemble method using the outputs of the various algorithm as input to a NN

## Dataset preparation
### UnderSampling
The dataset is imbalanced, most of the instances are of Cover_Type 1 and 2, we are however dealing with a large dataset, so even the most underrepresented cover type (Cover_Type 4 with 2747) still has a significant number of instances. We can therefore downsample the dataset as to have the same amount of instances for each cover type.

We will two undersampling strategies: 
* random undersampling
* NearMiss-3 undersampling

```python
from sklearn.utils import resample
from imblearn.under_sampling import NearMiss, RandomUnderSampler
from sklearn.preprocessing import LabelEncoder
```

### Logistic Regression
Let's start with a simple logistic regression, this will help us establish a baseline.

```python
def print_confusion_matrix(confusion_matrix_1, confusion_matrix_2, class_names, figsize = (16,7), fontsize=14):

    df_cm1 = pd.DataFrame(
        confusion_matrix_1, index=class_names, columns=class_names, 
    )
    df_cm2 = pd.DataFrame(
        confusion_matrix_2, index=class_names, columns=class_names, 
    )
    fig, axes = plt.subplots(1, 2, figsize=figsize)
    try:
        heatmap1 = sns.heatmap(df_cm1, annot=True, fmt="d", cmap="Blues", ax=axes[0])
        heatmap2 = sns.heatmap(df_cm2, annot=True, fmt="d", cmap="Blues", ax=axes[1])
    except ValueError:
        raise ValueError("Confusion matrix values must be integers.")
    heatmap1.yaxis.set_ticklabels(heatmap1.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=fontsize)
    heatmap1.xaxis.set_ticklabels(heatmap1.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=fontsize)
    heatmap2.yaxis.set_ticklabels(heatmap2.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=fontsize)
    heatmap2.xaxis.set_ticklabels(heatmap2.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=fontsize)
    axes[0].set_title('NearMiss3')
    axes[0].set_ylabel('True label')
    axes[0].set_xlabel('Predicted label')
    bottom, top = axes[0].get_ylim()
    axes[0].set_ylim(bottom + 0.5, top - 0.5)
    axes[1].set_title('Random')
    axes[1].set_ylabel('True label')
    axes[1].set_xlabel('Predicted label')
    bottom, top = axes[1].get_ylim()
    axes[1].set_ylim(bottom + 0.5, top - 0.5)
   
    
labels= sorted([x for x in data['Cover_Type'].unique()])
```

```python
from sklearn.linear_model import LogisticRegression
C_list = [1,20,50,100,150]
acc_LR_NM = []
acc_LR_RU = []
seed = 0
```

```python
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import concatenate
from tensorflow.keras.optimizers import Adam

layer_nodes_factors_model = {
    6: [2.0 ,1.7, 1.4, 1.1, 0.8, 0.5],
    5: [2.0, 1.7, 1.3, 1.0, 0.7],
    4: [2.0, 1.5, 1.0, 0.5],
    3: [1.5, 1.0, 0.5],
    2: [1.5, 0.5]
}
acc_DNN_NM = []
acc_DNN_RU = []
```

#### Using original features

```python
def build_model(inputs, nclasses, nb_features, nlayers, layer_nodes_factors, finalAct="softmax"):
        x = Dense(int(nb_features*layer_nodes_factors.get(nlayers)[0]))(inputs)
        x = Activation("relu")(x)
        x = BatchNormalization()(x)
        for n in range(1,nlayers):
            x = Dense(int(nb_features*layer_nodes_factors.get(nlayers)[n]), activation="relu")(x)
            x = BatchNormalization()(x)
        x = Dense(nclasses)(x)
        x = Activation(finalAct, name="CovType")(x)

        model = Model(inputs=ins, outputs=x)
        return model
```

```python
nb_instances, nb_features = X_RU_train.shape
lrate = 0.4
nb_epoch=30
n_list=[2,3,4,5,6]
best_score_RU=0
best_score_NM=0
```

```python
for nlayers in n_list:
    ins = Input(shape=(nb_features,))
    model = build_model(ins, nclasses, nb_features, nlayers, layer_nodes_factors_model)
    model_optimizer=Adam(lr=lrate, decay=lrate / nb_epoch)
    model.compile(optimizer=model_optimizer, loss="categorical_crossentropy", metrics=["accuracy"])
    model.fit(X_RU_train,y_RU_train_hot,batch_size=10,epochs=nb_epoch,validation_split=0.1)
    result = model.evaluate(X_RU_test, y_RU_test_hot)
    print(result)
    acc_DNN_RU.append(result[1])
    if(result[1]>best_score_RU):
        model.save('best_RU_model')
        best_score_RU = result[1]
    model = build_model(ins, nclasses, nb_features, nlayers, layer_nodes_factors_model)
    model_optimizer=Adam(lr=lrate, decay=lrate / nb_epoch)
    model.compile(optimizer=model_optimizer, loss="categorical_crossentropy", metrics=["accuracy"])
    model.fit(X_NM_train,y_NM_train_hot,batch_size=10,epochs=nb_epoch,validation_split=0.1)
    result = model.evaluate(X_NM_test, y_NM_test_hot)
    print(result)
    acc_DNN_NM.append(result[1])
    if(result[1]>best_score_NM):
        model.save('best_NM_model')
        best_score_NM = result[1]
```

## Conclusions
We see that generally Naive Bayes and AdaBoost don't seem to be adapted to this problem. For Naive Bayes I believe this to be related to the correlation between the features. It could be interesting to remove highly correlated features and see how this impacts the classifier. For AdaBoost the default base classifier is a Decision Tree, different base classifiers could have been tested and the learning rate could have been adjusted. Given the poor performances compared to the other topologies I chose to explore more topologies instead of optimising these two algorithms. 

Best results were different depending on the under sampling strategy. As mentioned, KNN worked very well with the Nearmiss dataset, this is related to the selection strategy of NearMiss3 which is based on neighbours selection. The Support Vector and Random Forest Classifiers gave excellent results for both datasets. 

The simple DNN gave average results, the optimisation however was far from done, in particular a smaller learning rate could have been tested. I believe that would have resulted in a significant improvement. 

The ensemble strategy obtained by stacking the different classifiers in a Neural Network gave excellent results as would have been expected, to get the best performance Naive Bayes and AdaBoost predictions were removed. It would be interesting to see how much the performance is affected by including them. 

Overall Random Forest and SVC took less time to optimise and gave similar results to the Artificial Neural Networks therefore hindering their performance. 

Generally the NearMiss dataset outperformed the Randomly undersampled dataset by more than 10%. In particular the majority classes (1,2 and 3) seem to suffer more from the Random selection of data to be removed. These classes probably had quite sparse features and the random undersampling might not capture the underlying shape of the distributions. On the other hand the opposit might also be true, with NearMiss specifically choosing the instances that best allow to separate the different classes. Saving both approaches and further testing them on the remaining data would be the best strategy.
